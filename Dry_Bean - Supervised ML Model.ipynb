**Applying Models**

```python
#Data Spliting

x= df.iloc[:,:-1].values  
y= df.iloc[:, -1].values 
```

```python
#train test split

from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0) 
```

```python
x_train
```

Data Pre-Processing with Sklearn using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)  
```

```python
x_train
```

```python
#Importing EvaluAtion Libraries

from sklearn import metrics
from sklearn.metrics import precision_score,accuracy_score,recall_score,confusion_matrix,f1_score
from sklearn.metrics import classification_report
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(random_state=0)
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
LR.score(x_test,y_test)
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

**K Nearest Neighbors**

```python
#training the model

from sklearn.neighbors import KNeighborsClassifier  
KNN= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
KNN.fit(x_train, y_train)
```

```python
#predicting

y_pred=KNN.predict(x_test)
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

```python
KNN.score(x_test,y_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

**Support vector machine**

```python
from sklearn.svm import SVC
SVM = SVC(kernel = 'linear', random_state = 0)
SVM.fit(x_train, y_train)
```

```python
y_pred=SVM.predict(x_test)
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

```python
SVM.score(x_test,y_test)
```

**Decision tree**

```python
from sklearn.tree import DecisionTreeClassifier
dtc_model = DecisionTreeClassifier()
dtc_model.fit(x_train,y_train)
```

```python
y_pred = dtc_model.predict(x_test)
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

**Evaluation of All Models**

```python
models = [LR, KNN, SVM, dtc_model]

evaluation_results = []
```

```python
for model in models:
    # Predict on the test dataset
    y_pred = model.predict(x_test)
    
    # Compute evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    # Append the results to the evaluation_results list
    evaluation_results.append({'Model': str(model), 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1})

# Create a DataFrame to display the results
df_evaluation = pd.DataFrame(evaluation_results)

df_evaluation
```

**CONCLUSION:**

* LR, KNN and SVM gives same efficency which is 92%.
* Decision tree gives the efficency 89% which is lower than all other models.
* All the algorithm gives almost similar Efficency.
* Best can be Logistic, KNN or SVM_linear.
